# Cleaning starts here

```python
to_normalize = ["Less than high school graduate", "High school graduate", "Some college or associate's degree", "Bachelor's degree or higher", "Employed_16_above"]
for i in to_normalize:
  df[i] = df[i]/df['Total Population']
```

```python
# Comobordities by 100 to match [0,1] percentage scale used
for col_name in comorbidity_df.columns:
  if col_name == "ZCTA5":
    continue
  df[col_name] = df[col_name]/100
```

```python
# Dropping rows with less than 1000 people
df = df[df['Total Population']>=1000]
```
